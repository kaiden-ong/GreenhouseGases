```python
import pandas as pd
import os

data_folder = "/data"

file_list = [f for f in os.listdir(data_folder) if f.endswith('.csv')]
dataframes = []
for file in file_list:
    df = pd.read_csv(os.path.join(data_folder, file))
    dataframes.append(df)

combined_data = pd.concat(dataframes, ignore_index=True)
```

```python
for i, df in enumerate(dataframes):
    print(f"File {file_list[i]} columns: {df.columns}")
```

```python
combined_data['start_time'] = pd.to_datetime(combined_data['start_time'], errors='coerce')
combined_data['end_time'] = pd.to_datetime(combined_data['end_time'], errors='coerce')
```

```python
combined_data.drop(columns=['created_date', 'modified_date'], inplace=True)
```

```python
combined_data['original_inventory_sector'] = combined_data['original_inventory_sector'].str.lower()
```

```python
combined_data.drop_duplicates(inplace=True)
```
